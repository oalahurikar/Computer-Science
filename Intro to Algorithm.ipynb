```python
A = [8,2,4,9,3,6]
def insertion_sort(array):
    for slot in range(1, len(array)): 
        value = array[slot]
        test_slot = slot - 1
        while test_slot > -1 and array[test_slot] > value:
            array[test_slot + 1] = array[test_slot]
            test_slot = test_slot - 1
        array[test_slot + 1] = value
    return array
```

# QuickSort

The main difference between quicksort and mergesort is in where the “heavy lifting” is done. In mergesort, the dividing step is trivial and all the work is in combining the merged sub-lists at the end. However, the opposite is true of quicksort, where the dividing step does all the work, and the combining step is trivial (in fact, non-existent!).

* **Divide**--Pick a pivot element (typically the last item in the array). All numbers lower than this value go to the left and all elements higher to the right. For quicksort, this step is commonly known as partitioning.
* **Conquer**--All elements to the left are fed recursively back into the algorithm, as are elements to the right.
* **Combine**--No need to do anything at all. At this point the data should be sorted.



```python
def quickSort(arr):
    less = []
    pivotList = []
    more = []
    arr_length = len(arr)
    if arr_length <= 1:
        return arr
    else:
        pivot = sorted([arr[0],arr[len(arr) // 2],arr[-1]])[1] #arr[0] # Change this line
        for i in arr:
            if i < pivot:
                less.append(i)
            elif i > pivot:
                more.append(i)
            else:
                pivotList.append(i)
#       Recursion       
        less = quickSort(less)
        more = quickSort(more)
        return less, pivotList, more
```

```python
C=[3,4,5,1,2,12,0,14,3,11]
quickSort(C)
```
